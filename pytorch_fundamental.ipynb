### Reshaping, stacking, squeezing and unsqueezing tensors(help in dealing with the issue related to shape)

* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory
as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side
(hstack)
* Squeeze - removes all `1' dimensions from a tensor
* Unsqueeze - add a `1' dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
import torch
x= torch.arange(0,10)
x,x.shape
```

```python
x_reshaped = x.reshape(10,1)
x_reshaped,x_reshaped.shape
```

```python
x_reshaped2 = x.reshape(1,10)
x_reshaped2,x_reshaped2.shape
```

```python
x_reshaped3 = x.reshape(2,5)
x_reshaped3,x_reshaped3.shape
```

```python
z = x.view(1,10)
z,z.shape
```

```python
# changes in z will also come in x
z[:,0]=5
z,x
```

```python
x_stacked = torch.stack([x,x,x,x],dim=1)
x_stacked
```

```python
x_stacked = torch.stack([x,x,x,x],dim=0)
x_stacked
```

```python
x_stacked = torch.vstack([x,x,x,x])
x_stacked
```

```python
x_stacked = torch.hstack([x,x,x,x])
x_stacked
```

```python
x_squeezed = x_reshaped2.squeeze()
x_squeezed,x_squeezed.shape
```

```python
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed,x_unsqueezed.shape
```

```python
# permute 
x_original = torch.rand(size=(224,224,3))#height,width,color_chennel
x_permuted = torch.permute(input=x_original,dims=(2,0,1))#color_chennel,height,width
x_permuted.shape
```

```python
x_original[0,0,0],x_permuted[0,0,0]
```

### random seed : reduce randomness

```python
randomA = torch.rand(3,4)
randomB = torch.rand(3,4)
randomA == randomB,randomA,randomB
```

```python
random_seed = 42 
torch.manual_seed(random_seed)
```

```python
torch.manual_seed(random_seed)
randomC = torch.rand(3,4)
torch.manual_seed(random_seed)
randomD = torch.rand(3,4)
randomC,randomD,randomC==randomD
```

### putting our tensor on gpu
- we use gpu for faster computation
- defalt device is cpu is comparatively weaker than gpu

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
tensor = torch.tensor([1,2,3],device = device)
tensor,tensor.device
```

```python
# anather mathod 
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

```python
# if tensor is on gpu then we will not be able to convert to numpy
tensor_on_cpu = tensor_on_gpu.cpu().numpy()
tensor_on_cpu,tensor_on_cpu.device
```
